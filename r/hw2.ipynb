<h3>Answer 4</h3>

```
(a) all2006[which(all2006$Prevailing_Wage_Job_Title == "Software Engineer"), ]
    1785 line items match this criterion
(b) all2006[grep("Software Engineer", all2006$Prevailing_Wage_Job_Title), ]
```

````r
all2006 <- read.csv("2006.csv",header=TRUE,as.is=TRUE)
all2006$Wage_Offered_From <- as.numeric(gsub("\\$","", all2006$Wage_Offered_From))
all2006$Prevailing_Wage_Amount <- as.numeric(gsub("\\$","", all2006$Prevailing_Wage_Amount))
all2006$Wage_Offered_To <- as.numeric(gsub("\\$","", all2006$Wage_Offered_To))
all2006 <- all2006[all2006$Wage_Per == 'Year', ]
all2006 <- all2006[all2006$Wage_Offered_From > 20000, ]
all2006 <- all2006[all2006$Prevailing_Wage_Amount > 200, ]

se2006 <- all2006[which(all2006$Prevailing_Wage_Job_Title == "Software Engineer"), ]
cat("(a) ")
cat(dim(se2006)[1])
cat(" line items match this creiteria\n")

se2006 <- all2006[grep("Software Engineer", all2006$Prevailing_Wage_Job_Title), ]
cat("(b) ")
cat(dim(se2006)[1])
cat(" line items match this creiteria\n")
````

````r
aList <- list(name = c("stock1", "stock2", "stock3", "stock4"), cap.usd = c(5,6,7,8))
myDf <- data.frame(aList, stringsAsFactors = FALSE)

# Answer
grepDf <- function (aPattern, aDataFrame) {    
    A <- function(x) {
        any(grepl(aPattern, x))
    }
    apply(aDataFrame, 1, A) # Applies to each element in all the rows
}

# Test Arguements
grepDf("stock3", myDf)
grepDf("stock1|stock3", myDf)
grepDf(8, myDf)
myDf[grepDf(8, myDf), ]
````

<h3>Answer 6</h3>

````r
countries <- unique(all2006$Alien_Citizenship_Country)
uniqueCountries <- length(countries)
cat("(a)", uniqueCountries, " unique countries")
````

````r
# declare vectors of means and standard deviations for each country
means <- vector(mode = "numeric", length = uniqueCountries)
sds <- vector(mode = "numeric", length = uniqueCountries)

for(i in 1:uniqueCountries) {
    filter <- all2006[which(all2006$Alien_Citizenship_Country == countries[i]),] # filter the data frame for use
    means[i] <- round(mean(filter$Wage_Offered_From, na.rm = TRUE), digits=4) # add elements to the correct index
    sds[i] <- round(sd(filter$Wage_Offered_From, na.rm = TRUE), digits=2)
}

countryList <- list(country = countries, mean = means, sd = sds) # make a list and the dataframe
df = data.frame(countryList, stringsAsFactors=FALSE)
df <- df[with(df, order(-mean)), ] # sort the dataframe
df <- df[1:10, ] # chose the top 10 and print the dataframe
cat("(b)\n")
df
````

````r
func <- function(x) {
    which(all2006$Alien_Citizenship_Country == x)
}
result <- lapply(countries, func)
attributes(result)$names <- countries # assign the correct name attribute to the indexes
cat("(c)")
result[['HONG KONG']] # answer preview
````
