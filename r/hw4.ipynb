## Answer 2

```r
YEAR <- 5
PAR <- 100
PRICE <- 99
COUPON <- 5

f <- function(y) {
    coupons <- rep(COUPON, YEAR)
    coupons[length(coupons)] <- coupons[length(coupons)] + PAR
    returnRate <- rep(y, length(coupons))
    minusValues <- coupons / (cumprod(1+returnRate))
    return (PRICE - sum(minusValues))
}

g <- function(y) {
    return (f(y)^2)
}
```

```r
guess <- 0.05
result <- nlm( g, p=guess )
```

```r
xMin <- result$estimate
f(xMin)
```

```r
x <- seq(0, 0.5, by=0.01)
plot(x=x, y=sapply(x, f), type="l", lwd=3, lty=1)
abline(h=0)
lines(x=x, y=sapply(x, g), type="l", lwd=3, col='red')
```
